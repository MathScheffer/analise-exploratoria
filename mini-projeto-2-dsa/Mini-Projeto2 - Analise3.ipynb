```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
import zipfile
clean_data_path = "dataset/autos.zip"

print(zipfile.ZipFile(clean_data_path).filelist)
zf = zipfile.ZipFile(clean_data_path)
zf.open('autos.csv')
df = pd.read_csv(zf.open('autos.csv'),encoding = "utf-8")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
df.columns
```

```python
ax = sns.barplot(data=df,y='price',x='fuelType',hue='gearbox')
ax.set_title('Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio')
ax.set_ylabel('Preço Médio',fontsize=16)
ax.set_xlabel('Tipo de Combustível',fontsize=18)
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
fig,axs = plt.subplots(figsize=(8,6))
ax = sns.barplot(data=df,y='powerPS',x='vehicleType',hue='gearbox')

ax.set_ylabel('Potência Média',fontsize=16)
ax.set_xlabel('Tipo de Veículo',fontsize=18)

ax.set_title('Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio')

plt.show()
```
