```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
import zipfile
clean_data_path = "dataset/autos.zip"

print(zipfile.ZipFile(clean_data_path).filelist)
zf = zipfile.ZipFile(clean_data_path)
zf.open('autos.csv')
df = pd.read_csv(zf.open('autos.csv'),encoding = "utf-8")
```

## Número de veículos pertencentes a cada marca

```python
sns.set_style('whitegrid')
ax = sns.catplot(data=df,y='brand',kind='count',height=7, palette='Reds_r', aspect=1.5)

plt.xlabel('Número de Vendas', fontsize=16)
plt.ylabel('Marca',fontsize=18)
plt.title('Vendas por Marca', fontsize=18)

plt.show()
```
